LangChain中常见的提示词模板类型包括：

## 基础模板类型

**PromptTemplate** - 最基础的字符串模板，支持变量替换：
```python
from langchain.prompts import PromptTemplate

template = "请回答关于{topic}的问题：{question}"
prompt = PromptTemplate(
    input_variables=["topic", "question"],
    template=template
)
```

**ChatPromptTemplate** - 用于聊天模型的对话格式模板：
```python
from langchain.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "你是一个专业的{role}"),
    ("human", "{input}")
])
```
## 消息类型模板

**SystemMessagePromptTemplate** - 系统角色消息模板

**HumanMessagePromptTemplate** - 用户消息模板  

**AIMessagePromptTemplate** - AI助手消息模板


## 特殊用途模板

**FewShotPromptTemplate** - 少样本学习模板，包含示例：
```python
examples = [
    {"input": "happy", "output": "sad"},
    {"input": "tall", "output": "short"}
]

few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix="给出反义词：",
    suffix="输入：{adjective}\n输出：",
    input_variables=["adjective"]
)
```

**FewShotChatMessagePromptTemplate** - 聊天格式的少样本模板

**PipelinePromptTemplate** - 组合多个模板的管道模板

## 选择性模板

**ConditionalPromptTemplate** - 根据条件选择不同模板
**LengthBasedExampleSelector** - 基于长度选择示例的模板

这些模板类型可以根据不同的应用场景灵活组合使用，比如RAG应用、对话系统、代码生成等，每种都有其特定的优势和适用范围。
